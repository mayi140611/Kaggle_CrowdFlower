```python
# %load gen_kfold.py

"""
__file__

    gen_kfold.py

__description__

    This file generates the StratifiedKFold indices which will be kept fixed in
    ALL the following model building parts.

__author__

    Chenglong Chen < c.chenglong@gmail.com >

"""

import sys
import dill
from sklearn.model_selection import StratifiedKFold
sys.path.append("../")
from param_config import config
import numpy as np
```

```python
## load data
with open(config.processed_train_data_path, "rb") as f:
    dfTrain = dill.load(f)
```

```python
list(zip(["relevance", "query"], ["median_relevance", "qid"]))
```

```python
skf = [0]*config.n_runs
for stratified_label,key in zip(["relevance", "query"], ["median_relevance", "qid"]):
    for run in range(config.n_runs):
        random_seed = 2015 + 1000 * (run+1)
#         skf[run] = StratifiedKFold(dfTrain[key], n_folds=config.n_folds,
#                                     shuffle=True, random_state=random_seed)
#         skf[run] = StratifiedKFold(n_splits=config.n_folds,shuffle=True, random_state=random_seed)
#         s.split(dfTrain[stratified_label],dfTrain[key])
        s = StratifiedKFold(n_splits=config.n_folds,shuffle=True, random_state=random_seed)
        l = []
        fold = 0
        for validInd, trainInd in s.split(np.arange(dfTrain.shape[0]),np.ones(dfTrain.shape[0])):
            print("================================")
            print("Index for run: %s, fold: %s" % (run+1, fold+1))
            print("Train (num = %s)" % len(trainInd))
            print(trainInd[:10])
            print("Valid (num = %s)" % len(validInd))
            print(validInd[:10])
            fold += 1
            l.append((validInd, trainInd))        
        skf[run]=l
    with open("%s/stratifiedKFold.%s1.pkl" % (config.data_folder, stratified_label), "wb") as f:
        dill.dump(skf, f)
```
